```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python


# Define the file path
file_path = "C:\\Users\\Ankita Singh\\assigmentfordataengineer\\5k_borrowers_data.csv"

# Load the data into a DataFrame
df = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to verify
df.head()
```

```python
# Print the column names
print(df.columns)
```

```python
# 1. Handle missing values
# Example: Fill missing numerical values with the mean
for column in df.select_dtypes(include=['float64', 'int64']).columns:
    df[column].fillna(df[column].mean(), inplace=True)
```

```python
# 2. Standardize text
# Example: Convert all text to lowercase
for column in df.select_dtypes(include=['object']).columns:
    df[column] = df[column].str.lower()
```

```python
df.head()
```

* Distribution of Credit Scores:

A histogram with a KDE (Kernel Density Estimate) plot to show the distribution of credit scores.

* Loan Amount by Loan Type:

A box plot to compare the loan amounts across different loan types.

* Interest Rate vs. Loan Amount:

A scatter plot to visualize the relationship between loan amounts and interest rates, with different loan types indicated by different colors.
* Delayed Payment by Gender:

A count plot to compare the frequency of delayed payments between different genders.
* EMI vs. Days Left to Pay Current EMI:

A scatter plot to show the relationship between EMI amounts and the days left to pay the current EMI, with delayed payments indicated by different colors.

```python
# 1. Distribution of Credit Scores
plt.figure(figsize=(10, 6))
sns.histplot(df['Credit Score'], bins=30, kde=True)
plt.title('Distribution of Credit Scores')
plt.xlabel('Credit Score')
plt.ylabel('Frequency')
plt.show()

```

```python
# 2. Loan Amount by Loan Type
plt.figure(figsize=(12, 6))
sns.boxplot(x='Loan Type', y='Loan Amount', data=df)
plt.title('Loan Amount by Loan Type')
plt.xlabel('Loan Type')
plt.ylabel('Loan Amount')
plt.xticks(rotation=45)
plt.show()

```

```python
# 3. Interest Rate vs. Loan Amount
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Loan Amount', y='Interest Rate', hue='Loan Type', data=df)
plt.title('Interest Rate vs. Loan Amount')
plt.xlabel('Loan Amount')
plt.ylabel('Interest Rate')
plt.show()

```

```python
# 4. Delayed Payment by Gender
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', hue='Delayed Payment', data=df)
plt.title('Delayed Payment by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()

```

```python
# 5. EMI vs. Days Left to Pay Current EMI
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Days Left to Pay Current EMI', y='EMI', hue='Delayed Payment', data=df)
plt.title('EMI vs. Days Left to Pay Current EMI')
plt.xlabel('Days Left to Pay Current EMI')
plt.ylabel('EMI')
plt.show()
```

```python
import pandas as pd
import mysql.connector

# Define the file path
file_path = "C:\\Users\\Ankita Singh\\assigmentfordataengineer\\5k_borrowers_data.csv"

# Load the data into a DataFrame
df = pd.read_csv(file_path)

# Handle missing values (simplified for demonstration)
df.fillna(method='ffill', inplace=True)

# Connect to MySQL database
conn = mysql.connector.connect(
    host='your_host',       # e.g., 'localhost'
    user='your_username',   # e.g., 'root'
    password='your_password',
    database='your_database'
)
cursor = conn.cursor()

# Create table
cursor.execute('''
CREATE TABLE IF NOT EXISTS borrowers (
    id INT AUTO_INCREMENT PRIMARY KEY,
    name VARCHAR(255),
    date_of_birth DATE,
    gender VARCHAR(10),
    marital_status VARCHAR(20),
    phone_number VARCHAR(20),
    email_address VARCHAR(255),
    mailing_address VARCHAR(255),
    language_preference VARCHAR(50),
    geographical_location VARCHAR(255),
    credit_score INT,
    loan_type VARCHAR(50),
    loan_amount FLOAT,
    loan_term INT,
    interest_rate FLOAT,
    loan_purpose VARCHAR(255),
    emi FLOAT,
    ip_address VARCHAR(50),
    geolocation VARCHAR(255),
    repayment_history TEXT,
    days_left_to_pay_current_emi INT,
    delayed_payment INT
)
''')

# Insert the DataFrame data into the database
for i, row in df.iterrows():
    cursor.execute('''
    INSERT INTO borrowers (name, date_of_birth, gender, marital_status, phone_number, email_address, mailing_address, language_preference, geographical_location, credit_score, loan_type, loan_amount, loan_term, interest_rate, loan_purpose, emi, ip_address, geolocation, repayment_history, days_left_to_pay_current_emi, delayed_payment)
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
    ''', tuple(row))

# Commit changes and close the connection
conn.commit()
conn.close()
```
